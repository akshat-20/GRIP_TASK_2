##### Importing needed packages

```python
import numpy as np
import pandas as pd
import sklearn.datasets as datasets
```

##### Reading data

```python
iris=datasets.load_iris()
#print(iris.DESCR)  description of dataset
#print(iris.data) 
#print(iris.feature_names) column names
#print(iris.target) target values
```

##### Data preparation

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df['target']=iris.target
df.head()
X=iris.data
y=iris.target
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df[iris.feature_names],df['target'],random_state=0,test_size=0.3)
```

```python
print("Shape of train dataset:")
print("X_train: ",X_train.shape)
print("y_train: ",y_train.shape)
print("Shape of test dataset:")
print("X_test: ",X_test.shape)
print("y_test: ",y_test.shape)
```

##### Training the decision tree algorithm

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree
```

```python
dtree.fit(X_train,y_train)
predTree=dtree.predict(X_test)
```

##### Graphical representation of tree

```python
# Import necessary libraries for graph viz
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

# Visualize the graph
dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data, feature_names=iris.feature_names,filled=True,rounded=True,special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

### Evaluation
we compare the actual values and predicted values to calculate the accuracy of a regression model. Evaluation metrics provide a key role in the development of a model, as it provides insight to areas that require improvement.

Mean absolute error: It is the mean of the absolute value of the errors.

```python
from sklearn import metrics
import matplotlib.pyplot as plt
print("Accuracy score: ",metrics.accuracy_score(predTree,y_test))
```
